```python
df_weather = pd.read_csv('data\Europe_weather.csv',sep=',') 
print(df_weather.shape)
df_weather.head()
```

```python
# delete residual columns in items
df_items.drop(['item_name','style','brand_id','group0_id','group1_id','group2_id','age','color','size'], axis = 1, inplace=True)
df_items.head()
```

```python
# delete residual columns in orders
df_orders.drop(['order_id', 'payment', 'department' , 'gift_quantity','unit_cogs', 'unit_rrp_vat_excl', 'channel','owner', 'CreatedAt' ], axis = 1, inplace=True)
df_orders.head()
```

```python
# delete residual columns in Weather
df_weather.drop(['city'], axis = 1, inplace=True)
df_weather.head()
```

```python
# Joining  Orders, items and weather
df_merged = df_orders.merge(df_items, on='item_code', how='left') 
df_merged.rename( columns = {'name':'brand_name'}, inplace = True)
df_merged = df_merged.merge(df_weather, on=['country_code','date'], how='left') 
```

```python
# Convert 'data' to date type
df_merged['date'] = pd.to_datetime(df_merged['date'])

# Generate different data from date
df_merged['month']      = df_merged['date'].dt.month
df_merged['year_month'] = df_merged['date'].dt.to_period('M')
df_merged['year_week']  = df_merged['date'].dt.strftime('%Y-%U')
df_merged['weekday']    = df_merged['date'].dt.dayofweek
df_merged['yearweek']   = df_merged['date'].dt.strftime('%U')
df_merged['yearday']    = df_merged['date'].dt.dayofyear
```
